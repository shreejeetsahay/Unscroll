```python
import pandas as pd
```

```python
exp_df = pd.read_csv("experience_sampling.csv")
```

```python
exp_df
```

```python
pinfo = pd.read_csv("participant_info.csv")
```

```python
pinfo
```

```python
exp_with_group = exp_df.merge(pinfo[["Participant ID", "Group"]], on="Participant ID", how="left", suffixes=("", "_pinfo"))
```

```python
exp_df1 = exp_with_group
```

```python
exp_df1.columns = ["pid", "day_no", "time_spent", "feeling", "content", "relaxed", "happy", "anxious", "worried", "tense", "Group"]
```

```python
exp_df1.drop(columns="feeling",inplace=True)
```

```python
exp_df1 = exp_df1.sort_values(by=["pid", "day_no"], ascending=[True, True])
```

```python
exp_df1
```

```python
import matplotlib.pyplot as plt

df = exp_df1

# Creating separate plots for each group
for group in df['Group'].unique():
    group_df = df[df['Group'] == group]
    plt.figure(figsize=(10, 6))
    for pid in group_df['pid'].unique():
        pid_data = group_df[group_df['pid'] == pid]
        plt.plot(pid_data['day_no'], pid_data['time_spent'], label=f'Participant {pid}', marker='o')
    # plt.title(f'Time Spent by Day for {group} Group')
    plt.xlabel('Day Number')
    plt.ylabel('Time Spent (hours)')
    plt.xticks(range(1,5))
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
exp_df1.columns
```

```python
exp_df1["esi_score"] = exp_df1["content"] + exp_df1["relaxed"] + exp_df1["happy"] - exp_df1["anxious"] - exp_df1["worried"] - exp_df1["tense"]
```

```python
exp_df1
```

```python
import matplotlib.pyplot as plt

df = exp_df1

# Creating separate plots for each group
for group in df['Group'].unique():
    group_df = df[df['Group'] == group]
    plt.figure(figsize=(10, 6))
    for pid in group_df['pid'].unique():
        pid_data = group_df[group_df['pid'] == pid]
        plt.plot(pid_data['day_no'], pid_data['esi_score'], label=f'Participant {pid}', marker='o')
    # plt.title(f'ESI score trends for {group} group')
    plt.xlabel('Day Number')
    plt.ylabel('ESI Score')
    plt.xticks(range(1,5))
    plt.legend()
    plt.grid(True)
    plt.show()
```
